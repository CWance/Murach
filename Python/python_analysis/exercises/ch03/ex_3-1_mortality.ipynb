# Exercise 3-1: Create some plots

```python
import pandas as pd
```

## Get the data

```python
mortality_data = pd.read_pickle('mortality_prepped.pkl')
mortality_data.head()
```

```python
mortality_wide = pd.read_pickle('mortality_wide.pkl')
mortality_wide.head()
```

## Visualize the data

```python
mortality_data.query('AgeGroup == "01-04 Years"').plot.scatter(x='Year', y='DeathRate')
```

```python
mortality_wide.plot()
```

```python
mortality_wide['15-19 Years'].plot(title = "Child Mortality Rates: 15 - 19 Years", ylabel = "Deaths per 100,000")
```

```python
mortality_wide.plot.area(legend = 'reverse',title = "Child Mortality Rates")
```

```python
mortality_wide.query('Year == 1900').plot.bar(title = "Child Mortality Rates of the year 1900",ylabel = "Deaths per 100,000")
```

```python
mortality_wide.query('Year == 2000').plot.bar(ylabel = "Deaths per 100,000",title = "Child Mortality Rates of the year 2000")
```

```python
mortality_data.pivot(index = 'Year', columns = 'AgeGroup', values = 'DeathRate').query('Year == 2000').plot.bar(title = "Child Mortality Rates", ylabel = "Deaths per 100,000")
```

```python
mortality_data.plot.hist(y ='DeathRate', bins = 15)
```

```python
mortality_data.plot.density(y = 'DeathRate', title = "Child Mortality Rates of the year 1900", grid = True)
```

```python
mortality_data.query('Year%25 == 0').groupby('Year')['DeathRate'].sum().plot.pie()
```

```python
mortality_wide.query('Year%25 == 0').plot.barh(title = "Child Mortality Rates of the year 1900", ylabel = "Child Death Rates")
```
